# 区域卷积神经网络R-CNN系列
除了单发多框检测以外，RCNN也是将深度模型应用于目标检测的开创性工作之一

这个notebook包含了快速RCNN，更快速RCNN和掩码RCNN

## RCNN
选取若干个提议区域（类似锚框的方法），并标注它们的类别和边界框(如偏移量).然后用CNN对每个提议区域进行前向传播以提取其特征。然后使用每个提议区域的特征来预测类别和边界框。

细节：
- 使用选择性搜索来选取多个高质量的提议区域，而不是锚框的遍历
- 使用预训练的CNN来提取特征
- 使用提议区域连同其标注的类别作为一个样本。训练多个SVM对目标进行分类，每个SVM判断样本属于哪个类别
- 将每个提议区域的特征联通其标注的边界框作为一个样本，训练线性回归模型来预测真实的边界框

## FAST RCNN
RCNN的性能瓶颈在于高质量的提议区域是有重叠的，这会导致对某个区域特征的重复计算，FAST RCNN仅在整张图像上执行卷积神经网络的前向传播。

细节：
- FASTRCNN用来提取特征的CNN的输入是整张图像而不是所谓的各个高质量提议区域，而且这个CNN会参加训练。
- 假设选择性搜索生成了$n$个提议区域，这些形状各异的提议区域在卷积神经网络的输出上分别标出了形状各异的兴趣区域。然后需要抽取出一样维度的特征以便于concatenation，所以FASTRCNN引入了兴趣区域汇聚层，抽取特征
- 通过全连接层转换输出形状
- 预测$n$个区域中每个区域的类别和边界框。老样子预测类别的时候我们使用softmax回归

## FASTER RCNN
FASTRNN需要在选择性搜索中生成大量的提议区域。FASTER RCNN将其替换成区域提议网络，从而减少提议区域的生成数量，并保证精度。

- 使用非极大抑制方法来减少提议区域的数量？

## Mask RCNN
基于FASTER RCNN 修改而来，将兴趣区域汇聚层替换成兴趣区域对齐层，使用双线性插值来保留特征图上面的空间信息，从而更适用于像素级预测

# 总结

- R-CNN 是最早最有名的基于锚框的和基于CNN的目标检测算法

- Fast/Faster R-CNN 持续提升性呢能

- Faster R-CNN 和Mask R-CNN实在追求高精度场景下的常用算法

# YOLO
- 主要特点是快，比SSD还快
- SSD 中的锚框大量地重叠，浪费了很多的计算
- YOLO将图片均匀分成$S\times S$ 个锚框
- 每个锚框预测$B$个边缘框
- 后续版本有持续的改进
